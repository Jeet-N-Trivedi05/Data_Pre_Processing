### Two methods of handling missing values 

1. imputation 
2. dropping

### Outlier analysis and filling missing values with median 

```python
fig , ax = plt.subplots(figsize=(8,4))
sns.distplot(df.salary)
```

```python
salary_gr_6 = (df['salary'] >= 600000)
df[salary_gr_6]
```

```python
fill_sal_median = df['salary'].median()
fill_sal_median
```

```python
df['salary'].fillna(fill_sal_median, inplace=True)
df.head()
```

```python
print(df.isna().sum())
```
