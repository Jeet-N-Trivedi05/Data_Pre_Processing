```python
import pandas as pd
```

```python
iris = pd.read_csv('/Users/jeettrivedi/Desktop/Data Science/Data Pre-Processing/Data/iris_data.csv')
```

## 1) Make correlation metrix and store it into a variable

```python
corr_iris = iris.corr()
# corr_iris
```

## 2) Unstack and sort the matrix in to a pandas series

```python
sorted_corr = corr_iris.unstack().sort_values(ascending=False)
# sorted_corr
```

## 3) Make dataframe of previously unstacked series

```python
corr_dataframe = pd.DataFrame(sorted_corr).reset_index()
# corr_dataframe
```

## 4) There will be some data with same name in both column remove them

```python
remove_same_value = corr_dataframe['level_0'] == corr_dataframe['level_1']
index_of_same_value = corr_dataframe[remove_same_value].index
```

```python
corr_dataframe.drop(index_of_same_value,inplace=True)
```
